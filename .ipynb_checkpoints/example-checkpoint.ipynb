```python
from GeoPrivacy.mechanism import *
from GeoPrivacy.metrics import *
from GeoPrivacy.processor import *
import geopandas as gpd
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
zones = gpd.read_file('Data/taxi_zones.shp')
zones = zones.to_crs("EPSG:4326")
```

```python
zones.head()
```

```python
zones['Centroid'] = zones['geometry'].apply(lambda x: x.centroid)
x_list = [(lon, lat) for (lon, lat) in zones['Centroid'].apply(lambda x: (x.xy[1][0], x.xy[0][0]))]
pi_list = [geometry.area for geometry in zones['geometry']]
dQ = geodesic_mile

epsilon = 0.5
delta = 1.1
```

```python
spanner = generate_spanner(x_list, dQ, delta)
```

```python
print('Number of edges in the original graph: ', len(x_list) * (len(x_list) - 1))
print('Number of edges in the generated graph: ', len(spanner.edges))
```

```python
p_matrix, pre_prob, post_prob = optql_graph_spanner(x_list, pi_list, spanner, dQ, epsilon=epsilon, delta=delta)
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(p_matrix, vmin=0, vmax=0.3)
plt.title('Stochastic Transition Matrix')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
trips = pd.read_csv('Data/trip_sample.csv', index_col=0)
```

```python
def encrypt_trip(pickup_index, dropoff_index, p_matrix):
    n = p_matrix.shape[0]
    pickup_index, dropoff_index = int(pickup_index), int(dropoff_index)
    encrypted_pickup_index = np.random.choice(range(n), p=p_matrix[pickup_index])
    encrypted_dropoff_index = np.random.choice(range(n), p=p_matrix[dropoff_index])
    return encrypted_pickup_index, encrypted_dropoff_index
```

```python
trips.apply(lambda x: encrypt_trip(x['PUlocationID'], x['DOlocationID'], p_matrix), axis=1)
```
